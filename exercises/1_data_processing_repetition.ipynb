```python
departments = set(df["Department"])
departments
```

```python
department_stats = []
for d in departments:
    dep_df = df[df["Department"]==d]
    average = dep_df["Salary_SEK"].mean().round(0)
    median = dep_df["Salary_SEK"].median()
    department_stats.append({"Department" : d, "Average salary" : float(average), "Median salary" : float(median)})

department_stats
```
